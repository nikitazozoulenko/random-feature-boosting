```python
import ast
import matplotlib.pyplot as plt
import numpy as np

def find_all_best_params(file_path):
    best_params_list = []
    with open(file_path, 'r') as file:
        for line in file:
            if line.startswith("Best params:"):
                params_str = line.split("Best params:")[1].strip()
                # Remove the 'modelClass' part to avoid syntax error
                params_str = params_str.replace("<class 'models.random_feature_representation_boosting.GradientRFRBoostClassifier'>", "'GradientRFRBoostClassifier'")
                params_dict = ast.literal_eval(params_str)
                best_params = {
                    'l2_cls': params_dict['l2_cls'],
                    'l2_ghat': params_dict['l2_ghat']
                }
                best_params_list.append(best_params)
    return best_params_list

def plot_histograms(best_params_list):
    l2_cls_values = [params['l2_cls'] for params in best_params_list]
    l2_ghat_values = [params['l2_ghat'] for params in best_params_list]

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.hist(l2_cls_values, bins=np.logspace(np.log10(min(l2_cls_values)), np.log10(max(l2_cls_values)), 20))
    plt.xscale('log')
    plt.title('Histogram of l2_cls')
    plt.xlabel('l2_cls')
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    plt.hist(l2_ghat_values, bins=np.logspace(np.log10(min(l2_ghat_values)), np.log10(max(l2_ghat_values)), 20))
    plt.xscale('log')
    plt.title('Histogram of l2_ghat')
    plt.xlabel('l2_ghat')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# Usage
file_path = '/home/nikita/Code/random-feature-boosting/testing.txt'
best_params_list = find_all_best_params(file_path)
plot_histograms(best_params_list)
```
